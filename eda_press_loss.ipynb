```python
import pylab
ecc = sorted(list(press_data['ecc'].unique()))
#pylab.rcParams['figure.figsize'] = (10,10)
for i in ecc:
    plt.scatter(press_data[press_data['ecc']==i]['flow_rate_gpm'],press_data[press_data['ecc']==i]['diff_p'],label=i,alpha=0.5)
    plt.xlabel('Flow Rate')
    plt.ylabel('DIfferential Pressure')
    plt.legend()
```
